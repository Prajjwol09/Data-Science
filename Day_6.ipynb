```python
# replacing empty values 

# replacing null values with number 130 

df = pd.read_csv('data.csv')

df.info()
```

```python
df.fillna(130, inplace = True)      # replaced the null values with 130 

df.info()       # resulted no null values
```

```python
# replace only for specified columns

df = pd.read_csv('data.csv')

df.info()
```

```python
df.fillna({'Calories': 130}, inplace = True)    # filled the null calories values with 130.

df.info()
```

```python
# replacing using Mean, Median and Mode 

mean_df = pd.read_csv('data.csv')

mean_df.info()
```

```python
mean_x = mean_df['Calories'].mean()

mean_df.fillna({'Calories': mean_x}, inplace = True)        #replaced the null calories column with the mean of it.

mean_df.info()
```

```python
df = pd.read_csv('data.csv')

df.info()
```

```python
mode_x = df['Calories'].mode()[0]       # accesses the first element of the mode series

mode_x

df.fillna({'Calories': mode_x}, inplace = True)

df.info()
```
