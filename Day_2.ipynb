```python
# positional arguments 

def impute_age(name, age):
    print('My name is ', name , ' and I am ' , age , ' years old.')

impute_age('Prajjwol Thapa', 21)
```

```python
# mixing positional and keyword arguments 

def mixing_args(dog, cat):
    print('I have a dog named as ', dog, 'and cat named as ', cat)

mixing_args('Khairey', cat='Millie')
```

```python
# passing different data types 

def my_function(fruits):
    for fruit in fruits:
        print(fruit)
        
fruit = ['Apple', 'Banana', 'Grapes']
my_function(fruit)
```

```python
# returning values 

def add_number(x,y):
    return x+y

add_number(2,3)
```

```python
# returning different data types 

def my_pets():
    return ['Dog', 'Cat', 'Lion']

pets = my_pets()

pets[2]
```

```python
# positional only arguments 

def pos_args(dog, /):
    print(dog)

pos_args('khaire')
```

```python
# keyword only arguments 

def key_args(*, cat, rat):
    print(cat, rat)

key_args(cat = 'Millie', rat='Chua')
```

```python
# combining positional only  and keyword only 

def combining(fname, lname, /, *, age):
    print('My name is ',fname, lname, 'and i am ', age, 'years old.')

combining('Prajjwol', 'Thapa', age = 21)
```

```python
# arbitrary arguments *args  allows a function to accept any number of positional arguments

def my_friends(*friends):
    print('My friend is a cricketer and his name is', friends[2])

my_friends('Niraj', 'Pramod', 'Prashant')
```

```python
# args with regular arguments 

def my_function(greeting, *names):
    for name in names:
        print(greeting, name)
my_function('Good Morning!', 'Prajjwol', 'Niraj', 'Pramod')
```

```python
# sum of any numbers using args

def addition(*numbers):
    total = 0
    for num in numbers:
        total += num
    return total

addition(10,20,30,40)
```

```python
# finding the max value

def max_number(*numbers):
    if len(numbers) == 0:
        return None

    max_num = numbers[0]
    for num in numbers:
        if num > max_num:
            max_num = num
    return max_num

print(max_number(10,20,30,100))
```

```python
# arbitary keyword arguments **kwargs

def my_kids(**kids):
    print('His lastname is', kids['lname'])

my_kids(fname = 'Prajjwol', lname = 'Thapa')

```

```python
# combining *args and *kwargs

def combo(Bname, *author, **coauthor):
    print('My fav book is', Bname, 'and its author is', author, 'also its coauthor is', coauthor)

combo('Python', 'Prajjwol','Pramod', coauthor= 'Niraj')
```

```python
# unpacking lists with * 

def my_function(a, b, c):
    return a+b+c

numbers = [1,2,3]
result = my_function(*numbers)
print(result)
```

```python
# local scope 

def l_Scope():
    x= 300
    print(x)

l_Scope()
```

```python
# function inside function

def l_Scope():
    x= 200
    def nested_Scope():
        print(x)
    nested_Scope()
    
l_Scope()
```

```python
# global scope 

x = 100

def global_function():
    print(x)

global_function()
```

```python
# naming variables 

x = 100 

def local_function():
    x = 200
    print(x)

local_function()      # it will first print the local value then the global value

print(x)
```

```python
# using global keyword 

def global_word():
    global x                  # makes the variable x adn its assigned value available globally
    x= 100

global_word()
print(x)
```

```python
# non local keyword 

def  non_local_1():
    x = 'Niraj'
    def non_local_2():
        nonlocal x             # the value of x belongs to the non_local_1 with the use of nonlocal
        x = 'Wassup'
    non_local_2()
    return x

print(non_local_1())
```

```python
# loops 

# the while loop 

num = 1 
while num <= 6:
    print(num)
    num += 1
```

```python
# the break statement 

num = 0

while num <= 7:
    print(num)
    if num == 5:
        break            # break condition stops the loop when num == 5
    num += 1
```

```python
# the continue statement 

num = 0
while num <=7:
    num +=1
    if num == 5:
        continue          # the continue statement skips that called iteration and starts other execution
    print(num)
```

```python
# for loops 

names = ['Prajjwol', 'Niraj','Bhups']
for name in names:
    print(name)        # prints each name in the list 
    print(name[2])      # print the 2nd index of each names
```

```python
# looping through a string
for letter in 'Ronaldo':
    print(letter)
    if letter == 'a':
        break
```

```python
# for loop with continue statement 

for letter in 'PrajjwolThapa':
    if letter == 'j':
        continue
    print(letter)
```

```python
# range function

for num in range(1,10):
    print(num)          # print len(num)-1
    if num == 5:
        break
```

```python
# nested loops

names = ['Prajjwol', 'Niraj', 'Pratik']
last_name = ['Thapa', 'Kafle','Adhikari']

for name in names:
    for lname in last_name:
        print(name, lname)
```

```python
# range of indexes 

fruits = ['Apple', 'Banana', 'Cherry', 'Kiwi','Melon','Mango']

print(fruits[0:3])      # prints the result upto the 2nd index meaning n-1

print(fruits[:4])       # starts from the beginning that is 0th index

print(fruits[3:])       # prints till the last index

print(fruits[:-4])      # the negative index starts from the last with -1
```

```python
# extending the list 

num_1 = [1,2,3,4,5]
num_2 = [6,7,8,9,10]

num_1.extend(num_2)     # joins the num_2 list with num_1 list
print(num_1)            # we can join list with any other objects like tuples, dictionaries
```
