```python
# merging 

# concat 

import numpy as np 
import pandas as pd 

# creating a dataframe
df1 = pd.DataFrame(
    {
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    },
    index= [0,1,2,3]
)
```

```python
df2 = pd.DataFrame(
    {
        "A": ["A4", "A5", "A6", "A7"],
        "B": ["B4", "B5", "B6", "B7"],
        "C": ["C4", "C5", "C6", "C7"],
        "D": ["D4", "D5", "D6", "D7"],
    },
    index=[4, 5, 6, 7],
)
```

```python
df3 = pd.DataFrame(
    {
        "A": ["A8", "A9", "A10", "A11"],
        "B": ["B8", "B9", "B10", "B11"],
        "C": ["C8", "C9", "C10", "C11"],
        "D": ["D8", "D9", "D10", "D11"],
    },
    index=[8, 9, 10, 11],
)
```

```python
frames = [df1, df2, df3]
```

```python
# concating the frames

# concat makes a full copy of data and iteratively using concat can create unnecessary copies. Thats why collect the objects in the list before concating.

result = pd.concat(frames)    
```

```python
result
```

```python
# joins 

df4 = pd.DataFrame(
    {
        "B": ["B2", "B3", "B6", "B7"],
        "D": ["D2", "D3", "D6", "D7"],
        "F": ["F2", "F3", "F6", "F7"],
    },
    index=[2, 3, 6, 7],
)
```

```python
result = pd.concat([df1, df4], axis= 1)
```

```python
result
```

```python
# inner join            # takes the intersection of the axis value.

result = pd.concat([df1, df4], axis= 1, join= 'inner')

result
```

```python
# reindexing

result = pd.concat([df1, df4], axis= 1).reindex(df1.index)

result          # excludes the row 4 and 5 because they are not in the df1. 
```

```python
# ignore_index ignores the overlapping indexes.

result = pd.concat([df1, df4], ignore_index= True)

result
```

```python
# creating a customers and transaction table 

customers = pd.DataFrame({
    "customer_id": [1, 2, 3, 4],
    "customer_name": ["Ram", "Sita", "Hari", "Gita"],
    "city": ["Kathmandu", "Pokhara", "Kathmandu", "Biratnagar"]
})


transactions = pd.DataFrame({
    "transaction_id": [101, 102, 103, 104, 105],
    "customer_id": [1, 2, 2, 5, 1],
    "amount": [500, 1200, 800, 400, 700]
})
```

```python
customers
```

```python
transactions
```

```python
# merging

#inner join 

merged_dataset = pd.merge(customers, transactions, on= 'customer_id', how= 'inner')

merged_dataset
```

```python
# left join 

left_join = pd.merge(customers, transactions, on= 'customer_id', how= 'left')

left_join
```

```python
# right join 

right_join = pd.merge(customers, transactions, on= 'customer_id', how='right')

right_join
```

```python
# concat --- stacking data row wise

new_customers = pd.DataFrame(
    {
    "customer_id": [5, 6],
    "customer_name": ["Nabin", "Asha"],
    "city": ["Lalitpur", "Bhaktapur"]
    }
)
```

```python
new_customers
```

```python
combined_customers = pd.concat([customers, new_customers], axis= 0, ignore_index= True)
```

```python
combined_customers
```

```python
# column wise concat 

customer_age = pd.DataFrame({
    'age': [25,30,35,28]
})

customer_with_age = pd.concat([customers, customer_age], axis=1)

customer_with_age
```

```python
# find total transaction amount per customer 

total_spent = transactions.groupby('customer_id')['amount'].sum().reset_index()

total_spent
```

```python
# merging with the customer table 

customer_spending = pd.merge(customers, total_spent, on='customer_id', how= 'left')

customer_spending
```

```python
# identify customers with no transactions 

customer_spending[customer_spending['amount'].isna()]       # here isna looks after the missing and empty values
```

```python
# replace nan amounts with 0 

customer_spending['amount'] = customer_spending['amount'].fillna(0)

customer_spending
```
