```python
import numpy as np 
```

```python
# vectorization-- converting the iterative statements into a vector based operation

x = [1,2,3,4,5]

y = [1,2,3,4,5]

z = np.add(x,y)     # here add is the unfunc, using this we can add without iterating the elements of array

print(z)
```

```python
# creating our own unfunc

def add(x,y):
    return x+y 

myadd = np.frompyfunc(add, 2, 1)        # func, inputs, outputs

print(myadd([1,2,3,4], [1,2,3,4]))
```

```python
# simple arithematic 

arr1 = np.array([10, 11, 12, 13, 14, 15])
arr2 = np.array([20, 21, 22, 23, 24, 25])

new_array_add = np.add(arr1, arr2)

new_array_add
```

```python
new_array_subtract = np.subtract(arr2, arr1)

new_array_subtract
```

```python
new_array_multiply = np.multiply(arr1, arr2)

new_array_multiply
```

```python
new_array_division = np.divide(arr1, arr2)

new_array_division
```

```python
new_Array_power = np.power(arr1, arr2)

new_Array_power
```

```python
new_Array_mod = np.mod(arr1, arr2)

new_Array_mod
```

```python
arr3 = [-10,-20,-30]

new_array_absolute = np.absolute(arr3)

new_array_absolute
```

```python
# summations 

arr1 = np.array([1, 2, 3])
arr2 = np.array([1, 2, 3])

new_array_sum = np.sum([arr1, arr2])        # this gives the sum of that array

new_array_sum
```

```python
# summations over an axis 

new_array_axis = np.sum([arr1, arr2], axis= 1)      # perform sum over an axis

new_array_axis
```

```python
# cumulative sum

new_arr_cumsum = np.cumsum(arr1)        # gives the cumulative sum

new_arr_cumsum
```

```python
# numpy products 

new_array_prod = np.prod(arr1)      # 1*2*3

new_array_prod
```

```python
new_arr_prods = np.prod([arr1, arr2])       # 1*2*3*1*2*3 meaning multiplies the number of both the array

new_arr_prods
```

```python
# cumulative product 

new_array_cumprod = np.cumprod(arr1)        # 1, 1*2, 1*2*3

new_array_cumprod
```

```python
# numpy trigonometric functions

x = np.sin(np.pi/2)

x
```

```python
# converting degrees into radians 

arr = np.array([90, 180, 270, 360])

x = np.deg2rad(arr)

x
```

```python
# radian to degrees

arr = np.array([np.pi/2, np.pi, 1.5*np.pi, 2*np.pi])

x = np.rad2deg(arr)

x

```

```python
# finding angles 

x = np.arcsin(1.0)      # returns the radian values 

y = np.rad2deg(x)

y
```

```python
# angles of each value in array 

arr = np.array([1, -1, 0.1])

x = np.arcsin(arr)

y = np.rad2deg(x)

y
```

```python
# hypotenuses 

base = 4 
perpendicular = 3 

height = np.hypot(base, perpendicular)       #calculates the height using pythagoras theorem

height
```

```python
# hyperbolic functions 

hyperbolic_fun = np.sinh(np.pi/2)

change_func = np.rad2deg(hyperbolic_fun)

change_func
```

```python
# finding angles 

x = np.arcsinh(1.0)

change_deg = np.rad2deg(x)

change_deg
```

```python
# rounding decimals 

# using trunc

arr = np.trunc([-3.1666, 3.6667])       # rounds to the closest zero.

print(arr)
```

```python
# using fix 

arr = np.fix([-3.1666, 3.667])

arr
```

```python
# using around 

arr = np.around(3.1666, 2)      # increases the preceding decimal by 1 if >=5 else do nothing.

arr
```

```python
# using floor 

arr = np.floor([-3.1666, 3.1999])       # rounds off decimal to nearest lowest integer.

arr
```

```python
# using ceil 

arr = np.ceil([3.1667, -3.1667])        # rounds off decuimal to nearest upper integer.

arr
```
